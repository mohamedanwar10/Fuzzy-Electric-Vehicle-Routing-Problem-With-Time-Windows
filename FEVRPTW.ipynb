```python

class EVRP():
    
    def __init__(self,points,n_iterations, ReadyTime, DueTime,ServiceTime,Demand,speed,Vheicles,Capacity,h):
        self.points= points                 #points that have x-axis and y-axis
        self.n_iterations= n_iterations     # The number of iteration for each arameters randomanization
        self.best_path = None               # store the best path for each parameters randomanization
        self.best_path_length = np.inf      # store the length for the best path for each parameters randomanization 
        self.n_points=points.shape[0]       #N ---> Number of nodes
        self.ReadyTime=ReadyTime            #Ej---> earlist allowed start time
        self.DueTime=DueTime                # Latest allowed start time 
        self.ServiceTime=ServiceTime        #Sj--->Servicetime 
        self.Demand=Demand                  # Dej---> demand 
        self.speed=speed                    # the speed for each car
        self.M=100                          # Big M
        self.BestTotalTime=np.inf           # The total time for each path 
        self.BestParameters={}              # The pest parameters that give the best path at all parameters 
        self.Customers=[False]*self.n_points # Boolean list that make all unserved customers with false by defualt 
        self.Vheicles=Vheicles        #V---> Number of Vhicles
        self.VheicleStatue=[False]*Vheicles   # Boolean list that make all unused vheicles with false by defualt 
        self.Capacity=Capacity        #Q---> Capacity of vheicles 
        self.StopTheAlgorithm=False  #will be true if we finish all the customers that achieve the constrients
        self.h=h                     #Energy consumption per 1 distence
        self.SOC=0                   #State Of Charging 
        #The dictionary to store all vheicles data
        self.Vheicle={'id':None,'capacity':self.Capacity,'statue':False,'charging':100,'path':None,'path length':0,'total time':0}
        self.EarlyPenalityCost=np.inf   #A penality cost if the vheicle reach before the Ready time for the customer
        self.LatePenalityCost=np.inf     #A penality cost if the vheicle reach after the Due time for the customer
        self.FuzzyDemand=self.Set_Fuzzy_Demand()  #Set the fuzzy demand for each customer
        self.Objective_Function_Value=np.inf  #Variable to catch the path that has the minimum objective function for each parameter
        self.VheiclesGroup= []     #Function to store all vheicles informations
      
    #Function that calculate distance between two points  
    @staticmethod
    def distance(x,y):
        return np.sqrt(np.sum((x - y)**2))

    
    
    
    #Function that Initialize the Phermone for each point
    def Initialize_Phermone(self):
        return np.ones((self.n_points, self.n_points))
    
    
    #Function that set the fuzzy demand for each customer : initialize minimum and maximum and most likely demand
    def Set_Fuzzy_Demand(self):
        fuzzy_demand={}
        for i in range(len(self.Demand)):
            #For each customer we cannot sure the customer's demand 
            #So we simulate a probably minimum demand and maximum demand to ensure that:
            #All maximum and minimum possible demand for each customer don't exceed the vheicle's capacity 
            min_demand=self.Demand[i]*0.5
            most_likely_demand=self.Demand[i]
            max_demand=self.Demand[i]*1.2
            
            fuzzy_demand[i]=(min_demand,most_likely_demand,max_demand)
            
        return fuzzy_demand      
    
    
    
    
    
    
    # Function to check if the total fuzzy demand exceeds vehicle capacity or no
    def check_fuzzy_demand(self, path, capacity):
        total_min = 0
        total_most_likely = 0
        total_max = 0

        
        for i in path:
            total_min += self.FuzzyDemand[i][0]
            total_most_likely +=self.FuzzyDemand[i][1]
            total_max += self.FuzzyDemand[i][2]
            
   
        if total_max <= capacity:
            return True

        elif total_min > capacity:
            return False
            
        else:

            # Calculate the degree of membership for the total demand
            total_demand = np.array([total_min, total_most_likely, total_max])
            capacity_array = np.array([0, capacity, capacity])
            membership = fuzz.trimf(total_demand, capacity_array)
            return membership[1] > 0.5  # Check if the most likely demand is within capacity


        
        
    def Handling_Customers(self,path):
        while not obj.check_fuzzy_demand(path,200):
            indx=0
            mx=0
            for i in path:
                if obj.FuzzyDemand[i][1]>mx:
                    mx=obj.FuzzyDemand[i][1]
                    indx=i
            path.remove(indx)
         
        #Calculate the informations
        path_length=0
        capacity=200
        SOC=100
        Zij_consumption=0
        total_time=0
        EarlyPenalityCost=0
        LatePenalityCost=0
        for i in range (len(path)-1):
            #Handling patth length
            path_length+=self.distance(self.points[path[i]],self.points[path[i+1]])
            
            #Handling the charging
            Zij_consumption +=self.h*self.distance(self.points[path[i]],self.points[path[i+1]])*100
                        
            LP_Cj=1             #late penality
            EP_Cj=1             #early penality
            if total_time + self.distance(self.points[path[i]],self.points[path[i+1]])/self.speed >self.DueTime[path[i+1]]:
                #Late penality cost
                LatePenalityCost= LP_Cj*self.M  # Add early penalty cost if arrival is before the earliest allowed start time


            if total_time + self.distance(self.points[path[i]],self.points[path[i+1]])/self.speed +self.ServiceTime[path[i+1]] < self.ReadyTime[path[i+1]]:   
               #arly penality cost
               EarlyPenalityCost += EP_Cj*self.M  # Add late penalty cost if arrival is after the latest allowed start time
                
            total_time += self.distance(self.points[path[i]],self.points[path[i+1]])/self.speed +self.ServiceTime[path[i+1]]
            
        SOC-= Zij_consumption   
        
        for customer in path:
            capacity-=self.Demand[customer]
        
        mn_obj =Zij_consumption*0.99+EarlyPenalityCost*0.002+LatePenalityCost*0.008
        
        return path,path_length,capacity,SOC,total_time,LatePenalityCost,EarlyPenalityCost,mn_obj
        
        
        
        
        
    # Function that plot the served and unserved customer and best path    
    def Plot_Best_Path(self, best_path):
        served_customers = np.where(self.Customers)[0]
        fig = plt.figure(figsize=(8, 6))
        ax = fig.add_subplot(111)
        
        # Plot all points in red
        ax.scatter(self.points[:, 0], self.points[:, 1], c='r', marker='o', label='Unserved Customers')

        # Highlight served customers in yellow
        ax.scatter(self.points[served_customers, 0], self.points[served_customers, 1], c='y', marker='o', label='Served Customers')

        # Plot the best path
        if len(best_path) < 2:
            print("Error: The best path should have at least 2 points.")
            self.StopTheAlgorithm = True
            return

        for i in range(len(best_path) - 1):
            if best_path[i] < len(self.points) and best_path[i+1] < len(self.points):
                ax.plot([self.points[best_path[i], 0], self.points[best_path[i+1], 0]],
                        [self.points[best_path[i], 1], self.points[best_path[i+1], 1]],
                        c='g', linestyle='-', linewidth=2, marker='o')
            else:
                print(f"Error: Index out of range for best path at index {i} or {i+1}.")
                return

        ax.plot([self.points[best_path[0], 0], self.points[best_path[-1], 0]],
                [self.points[best_path[0], 1], self.points[best_path[-1], 1]],
                c='g', linestyle='-', linewidth=2, marker='o')

        # Set labels and show plot
        ax.set_xlabel('X Label')
        ax.set_ylabel('Y Label')
        ax.legend()
        plt.show()
    
    
    
    
    #Function that print the best path results 
    def Show_Results(self):
        print("The best path is :")
        print(self.best_path)
        print("The length of this path is:") 
        print(self.best_path_length)
        print("The total time for this path is: ")
        print(self.BestTotalTime)
        print("The best parameters are: ")
        print(self.BestParameters)
        print("The best state of charging is:")
        print(self.SOC)
        print(f"The remain capacity is:")
        print(self.Capacity)
        print("The objective function value is:")
        print(self.Objective_Function_Value)
        print(f"The penality late cost is:{self.LatePenalityCost} and penality early cost is: {self.EarlyPenalityCost}")  
        self.Plot_Best_Path(self.best_path)

        
        
        

    
    
    # Ant colony optimization function 
    def ant_colony_optimization(self):
        #If there is a customer that not be served
        if not self.StopTheAlgorithm: 
            self.best_path = None
            self.best_path_length = np.inf
            self.BestTotalTime=np.inf
            self.SOC=0
            self.EarlyPenalityCost=np.inf
            self.LatePenalityCost=np.inf
            self.Capacity=200
            self.Objective_Function_Value=np.inf
            
            #The next loop to select the best parameters
            parameters={}
            for i in range(4):
                #Define the parameters
                evaporation_rate=np.round(min(abs(np.random.rand()-0.5),abs(np.random.rand()-0.9)),3)
                alpha=np.round(np.random.randint(0,4),3)
                beta=np.round(np.random.randint(1,5),3)
                Q=np.round(np.random.randint(1,6),3)
                ants=np.round(np.random.randint(1,25),3)     #Random the number of ants
                
             
                #Store all parameters 
                parameters['parameters '+str(i+1)]={'alpha':alpha,'beta':beta,'evaporation_rate':evaporation_rate,'Q':Q,'Ants':ants}
                pheromone = self.Initialize_Phermone()
                BestIterationPath=None
                BestIterationPathLength = np.inf
                BestTotalTime2=np.inf
                BestStatueOfCharging=0
                BestEarlyPenalityCost=0
                BestLatePenalityCost=0
                BestCapacity=0
                BestMinObjective=np.inf
                print('----------------------------------------------------------------------------------------------------')
                print(f"The values for  the parameters {parameters['parameters '+str(i+1)]}") 
                print('----------------------------------------------------------------------------------------------------')

                for iteration in range(self.n_iterations):
                    paths = []             
                    path_lengths = [] 
                
                    
                    for ant in range(ants):     
                        visited = self.Customers.copy()    
                        current_point = 0          
                        visited[current_point] = True                
                        path = [current_point]
                        path_length = 0
                        total_time=0         
                        StatueOfCharging=100 
                        LatePenalityCost=0 
                        EarlyPenalityCost=0
                        capacity=200

                        while False in visited:                               
                            unvisited = np.where(np.logical_not(visited))[0]    
                            probabilities = np.zeros(len(unvisited))              
                            
                            #filter the nodes that achieve constriants  
                            for i, unvisited_point in enumerate(unvisited):          
                                next_point = unvisited_point
                                ### Constriants handling ###
                                
                                #Initialize travel time 
                                Travel_time =self.distance(self.points[current_point],self.points[next_point])/self.speed
                                
                                # The charging consumption to travel from the current node and the next node 
                                Zij_consumption =self.h*self.distance(self.points[current_point],self.points[next_point])*100
                                
                                # The charging consumption to travel from the depot and the next point 
                                Zq_consumption =self.h*self.distance(self.points[0],self.points[next_point])*100
                                                  
                                if (total_time + Travel_time <= self.DueTime[next_point] 
                                and total_time + Travel_time + self.ServiceTime[next_point] >= self.ReadyTime[next_point] 
                                and StatueOfCharging>Zij_consumption 
                                and StatueOfCharging-Zij_consumption>Zq_consumption):
                                    # The action is:
                                    probabilities[i] = (pheromone[current_point, next_point] ** alpha) / (self.distance(self.points[current_point],self.points[next_point]) ** beta)
                                         
                                else:
                                    # Apply penalty costs for violating time window constraints
                                    probabilities[i] =0  # Setting the probability to 0 if the time window constraints are violated
            
                                    LP_Cj=1             #late penality
                                    EP_Cj=1             #early penality
                                    if total_time + self.distance(self.points[current_point],self.points[next_point])/self.speed >self.DueTime[next_point]:
                                       #Late penality cost
                                        LatePenalityCost= LP_Cj*self.M  # Add early penalty cost if arrival is before the earliest allowed start time
                    
                                    if total_time + self.distance(self.points[current_point],self.points[next_point])/self.speed +self.ServiceTime[next_point] < self.ReadyTime[next_point]:   
                                        #Early penality cost
                                        EarlyPenalityCost += EP_Cj*self.M  # Add late penalty cost if arrival is after the latest allowed start time
                                        
                    



                            if np.sum(probabilities) == 0:
                                if len(path)>1:
                                    Zij_consumption =self.h*self.distance(self.points[0],self.points[current_point])*100
                                    path_length += self.distance(self.points[current_point],self.points[0])
                                    StatueOfCharging-=Zij_consumption
                                    path.append(0)
                                break  # Exit the loop if no valid next point is found
                            else:
                                probabilities /= np.sum(probabilities)  # Normalize the probabilities
                                next_point = np.random.choice(unvisited, p=probabilities)        
                             # Update the total time considering the next point and its service time   
                            total_time += self.distance(self.points[current_point],self.points[next_point])/self.speed +self.ServiceTime[next_point]

                            path.append(next_point)
                            StatueOfCharging-=self.h*self.distance(self.points[current_point],self.points[next_point])*100
                            path_length += self.distance(self.points[current_point],self.points[next_point])
                            visited[next_point] = True
                            current_point = next_point
                            capacity-=self.Demand[current_point]


                        
                        if path[-1]!=0:
                            path.append(0)
                            path_length += self.distance(self.points[path[-1]],self.points[0])
                            Zij_consumption =self.h*self.distance(self.points[0],self.points[path[-1]])*100
                            path_length += self.distance(self.points[path[-1]],self.points[0])
                            StatueOfCharging-=Zij_consumption
                        paths.append(path)
                        path_lengths.append(path_length)
                         
                        # The objective function 
                        consumption=(100 - StatueOfCharging)
                        mn_obj =consumption*0.99+EarlyPenalityCost*0.002+LatePenalityCost*0.008

                        
                        if BestMinObjective > mn_obj :        
                            BestIterationPath = path
                            BestIterationPathLength = path_length
                            BestTotalTime2=total_time
                            BestStatueOfCharging=StatueOfCharging
                            BestEarlyPenalityCost=EarlyPenalityCost
                            BestLatePenalityCost=LatePenalityCost
                            BestCapacity=capacity
                            BestMinObjective=mn_obj
                            

               
                    pheromone *= (1 - evaporation_rate)

                    for path, path_length in zip(paths, path_lengths):
                        for i in range(len(path) - 1):
                            pheromone[path[i], path[i + 1]] += Q / path_length
                            
                    if BestIterationPath !=[0,0] and  BestIterationPath !=[0]:
                        print(f"The best path after the iteration number {iteration +1}")    
                        print(BestIterationPath)
                        print("The length is: ")
                        print(BestIterationPathLength)
                        print("Total time is: ")
                        print(BestTotalTime2)
                        print("The statue of charging is :")
                        print(BestStatueOfCharging)
                        print(f"The penality late cost is:{BestLatePenalityCost} and penality early cost is: {BestEarlyPenalityCost}")
                        print("The remain capacity is: ")
                        print(BestCapacity)
                        print("The objective function value is:",BestMinObjective)
                        print(" ")
                    else:
                        self.StopTheAlgorithm=True
                        self.best_path=[0,0]
                        return 'All cutomers that achieve the constriants are served !'



                ##Put the path that achieve a fuzzy demand here 
                if not self.check_fuzzy_demand(BestIterationPath, capacity=200):
                    BestIterationPath,BestIterationPathLength,BestCapacity,BestStatueOfCharging,BestTotalTime2,LatePenalityCost,EarlyPenalityCost,BestMinObjective=self.Handling_Customers(BestIterationPath)
                    
                self.Plot_Best_Path(BestIterationPath)
                
                if self.Objective_Function_Value > BestMinObjective:  
                    self.best_path = BestIterationPath
                    self.best_path_length = BestIterationPathLength
                    self.BestTotalTime= BestTotalTime2
                    self.SOC=BestStatueOfCharging
                    self.LatePenalityCost=LatePenalityCost
                    self.EarlyPenalityCost=EarlyPenalityCost
                    self.Capacity=BestCapacity
                    self.Objective_Function_Value=BestMinObjective
                    self.BestParameters=parameters['parameters '+str(i+1)]={'alpha':alpha,'beta':beta,'evaporation_rate':evaporation_rate,'Q':Q,'Ants':ants}
           
            
            obj.Show_Results()  
            
            
        else :
            return 'All cutomers that achieve the constriants are served !'
        




    def Distribute_Vheicles(self) :
        Unserved_Vheicles=self.Vheicle.copy()
        
        vheicles = self.VheicleStatue.count(True)
        while vheicles<self.Vheicles and not all(self.Customers):
            # print the vheicle
            print(f"For vheicle {vheicles+1}")
            
            self.ant_colony_optimization()

            
            if  len(self.best_path)==2 or len(self.best_path)==1:          
                remain_customers=np.where(np.logical_not(self.Customers))[0]
                indx=0
                if not remain_customers[0]:
                    indx+=1
                remain_customers=remain_customers[indx:].copy()
                self.Customers[0]=True
                #self.Customers[-1]=True
                if len(remain_customers)>0:
                    print(f"The remain customers are: {remain_customers}")
                    for customer in remain_customers:
                        print(f"For vheicle {vheicles+1}")
                        self.best_path=[0,customer,0]
                        self.VheicleStatue[vheicles]=True
                        self.Customers[customer]=True
                        print("The path is:")
                        print(self.best_path)
                        print(f"The total length is :")
                        print(self.distance(points[0],self.points[customer])*2)
                        print(f"The total time is :")
                        print((self.distance(points[0],self.points[customer])/self.speed)*2 + ServiceTime[customer])     
                        print(f"The remain capacity is :")
                        print(self.Capacity-self.Demand[customer])
                        print(f"The statue of charging is :")
                        Zij_consumption =(self.h*self.distance(self.points[0],self.points[customer])*100)*2
                        print(100-Zij_consumption)
                        self.Vheicle['statue']=True
                        self.Vheicle['capacity']=self.Capacity-self.Demand[customer]
                        self.Vheicle['id']=vheicles+1
                        self.Vheicle['path']=self.best_path
                        self.Vheicle['path length']=self.distance(points[0],self.points[customer])*2
                        self.Vheicle['total time']=(self.distance(points[0],self.points[customer])/self.speed)*2 + ServiceTime[customer]
                        self.Vheicle['charging']=100-Zij_consumption
                        self.VheiclesGroup.append(self.Vheicle.copy())
                        self.Plot_Best_Path(self.best_path)
                        vheicles+=1 
                        if vheicles>self.Vheicles:
                            print("We launched all vehicles !!")
                            break
                        print(f"We served {self.Customers.count(True)} customers")
                        print(" ")
                        print("**************************************************************************")
                        print(" ")
                self.Plot_Best_Path([0,0,0])
                break
            else:    
                for customer in self.best_path:
                    if customer==0 and self.Customers.count(True)==self.n_points-1:
                        self.Customers[0]=True
                    else:      
                        self.Customers[customer]=True    
                    
   
            # The vheicle is in the service
            
            self.Vheicle['id'] = vheicles + 1
            self.Vheicle['capacity'] = self.Capacity
            self.Vheicle['statue'] = True
            self.Vheicle['charging'] = self.SOC
            self.Vheicle['path'] = self.best_path
            self.Vheicle['path length'] = self.best_path_length
            self.Vheicle['total time'] = self.BestTotalTime
            self.VheiclesGroup.append(self.Vheicle.copy())
            vheicles+=1  # Increement to move to the next vheicle 
            if vheicles>self.Vheicles:
                print("We launched all vehicles !!")
                break
            print(f"We served {self.Customers.count(True)} customers")
            print(" ")
            print("**************************************************************************")
            print(" ")
            
        vheicles+=1 
        while vheicles <=self.Vheicles:
            Unserved_Vheicles['id']=vheicles
            self.VheiclesGroup.append(Unserved_Vheicles.copy()) 
            vheicles+=1
        
                
        
        
        
        
        
        
        
        
        
    
  #The end of the Class 
samples=101
points =np.array(np.matrix(nodes.iloc[:samples,[1,2]]))

ReadyTime=np.array(np.matrix(nodes['READY_TIME']))
ReadyTime = ReadyTime[0, :samples]
ReadyTime.reshape(1, samples)


DueTime=np.array(np.matrix(nodes['DUE_DATE']))
DueTime=DueTime[0, :samples]
DueTime.reshape(1, samples)


ServiceTime=np.array(np.matrix(nodes['SERVICE_TIME']))
ServiceTime=ServiceTime[0, :samples]
ServiceTime.reshape(1, samples)

Demand=np.array(np.matrix(nodes['DEMAND']))
Demand=Demand[0, :samples]
Demand.reshape(1, samples)


obj = EVRP(points=points,n_iterations=5,ReadyTime=ReadyTime, DueTime=DueTime, ServiceTime=ServiceTime, Demand=Demand,speed=20,Vheicles=25,Capacity=200,h=0.005)
obj.Distribute_Vheicles()

```
